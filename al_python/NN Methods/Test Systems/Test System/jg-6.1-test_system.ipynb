```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math
import tensorflow as tf
from tensorflow.python.framework import ops
```

```python
d_t = 0.01
d_t_decimal = str(d_t)[::-1].find('.')
steps = 6000
```

```python
class PID:
    def __init__(self, Kp=0, Ki=0, Kd=0):
        self.Kp = Kp
        self.Ki = Ki
        self.Kd = Kd
#         self.I_max = 20
        self.e = 0
        self.t = 0
        self.e_prev = 0
        self.t_prev = 0
        self.P = 0
        self.I = 0
        self.D = 0
        self.u = 0
        
    def set_state(self, e, t):
        self.e = e
        self.t = t
        
    def update(self):

        delta_time = self.t - self.t_prev
        delta_error = self.e - self.e_prev
        
        self.P = self.e
        
        self.I += self.e * delta_time
        
        if delta_time > 0:
            self.D = delta_error / delta_time
        else:
            self.D = 0

        self.t_prev = self.t
        self.e_prev = self.e

        self.u = self.Kp * self.P + self.Ki * self.I + self.Kd * self.D
        return np.float32(self.u)
        
        
    def gen_u(self, Kp, Ki, Kd):
        self.Kp = Kp
        self.Ki = Ki
        self.Kd = Kd
        
        delta_time = self.t - self.t_prev
        delta_error = self.e - self.e_prev
        
        P = self.e
        
        I = self.I + self.e * delta_time

        if delta_time > 0:
            D = delta_error / delta_time
        else:
            D = 0

        u = Kp * P + Ki * I + Kd * D
        return np.float32(u)
        
    def d_gen_u(self, Kp, Ki, Kd):        
        delta_time = self.t - self.t_prev
        delta_error = self.e - self.e_prev
        
        P = self.e
        
        I = self.I + self.e * delta_time

        if delta_time > 0:
            D = delta_error / delta_time
        else:
            D = 0
        return np.float32(P), np.float32(I), np.float32(D)
    
    def reset(self):
        self.e = 0
        self.t = 0
        self.e_prev = 0
        self.t_prev = 0
        self.P = 0
        self.I = 0
        self.D = 0
        self.u = 0
        
    def gain_from_FOPDT(self, K, tau, T):
        #658.PDF
        a1=1.086
        b1=-0.869
        a2=0.740
        b2=-0.130
        a3=0.348
        b3=0.914
        
        Kp = a1/K*(T/tau)**b1
        Ki = tau/(a2+b2*(T/tau))
        Kd = a3*tau*(T/tau)**b3
        
        self.Kp = Kp
        self.Ki = Ki
        self.Kd = Kd
```

```python
def plant_process(y, t, u):
    K = 0.01
    tau = 1.0
    T = 0.2
    
    t_delay = round(t-T, d_t_decimal)
    if (t_delay<= 0) or (t_delay<0.0):
        u_delay = u[0]
    else:
        step_t = int(t_delay/d_t)-1
        if step_t >= steps:
            u_delay = u[-1]
        else: 
            u_delay = u[step_t]  
        
    dydt = (-y + K*u_delay)/tau
    return dydt

def plant(y_0, t1, t2, u):
    ts = [t1, t2]
    return odeint(plant_process, y_0, ts, args=(u,))[1]
```

```python
interval = 500
r_steps = steps//interval

r_sq = np.zeros(steps)
up = 0

for i in range(r_steps):
    if i < r_steps-1:
        r_sq[i * interval:(i+1)*interval] = up
    else:
        r_sq[i * interval:] = up
    if up > 0:
        up = -1
    else:
        up = 1

t = np.linspace(0, d_t*steps, steps)
r_sin = np.sin(t)

def gen_rand_sig(t):
    N = 1000
    x = np.random.normal(scale=10.0, size=len(t))
    return np.convolve(x, np.ones((N,))/N)[(N-1):] 
    
r_rand = gen_rand_sig(t)
plt.plot(t, r_rand)

r_combo = np.append(r_sin[:steps//2], r_sq[steps//2:])
```

### Plant control with neural network

```python
pid = PID(1, 1, 1)
pid.gain_from_FOPDT(0.01, 1.0, 0.2)


np_gen_u = np.vectorize(pid.gen_u)
np_d_gen_u = np.vectorize(pid.d_gen_u)

def tf_d_gen_u(Kp, Ki, Kd,name=None):
    with tf.name_scope(name, "d_gen_u", [Kp, Ki, Kd]) as name:
        y = tf.py_func(np_d_gen_u,
                        [Kp, Ki, Kd],
                        (tf.float32, tf.float32, tf.float32),
                        name=name,
                        stateful=False)
        return y

def py_func(func, inp, Tout, stateful=True, name=None, grad=None):

    # Need to generate a unique name to avoid duplicates:
    rnd_name = 'PyFuncGrad' + str(np.random.randint(0, 1E+8))

    tf.RegisterGradient(rnd_name)(grad)  # see _MySquareGrad for grad example
    g = tf.get_default_graph()
    with g.gradient_override_map({"PyFunc": rnd_name}):
        return tf.py_func(func, inp, Tout, stateful=stateful, name=name)
    
def gen_ugrad(op, grad):
    Kp, Ki, Kd = op.inputs
    
    dudKp, dudKi, dudKd = tf_d_gen_u(Kp, Ki, Kd)
    return grad * dudKp, grad * dudKi, grad * dudKd


def tf_gen_u(Kp, Ki, Kd, name=None):

    with tf.name_scope(name, "gen_u", [Kp, Ki, Kd]) as name:
        y = py_func(np_gen_u,
                        [Kp, Ki, Kd],
                        [tf.float32],
                        name=name,
                        grad=gen_ugrad)  # <-- here's the call to the gradient
        return y[0]

#Train pid nn to output decent values first

r_delay = 5
u_delay = 5
y_p_delay = 5

graph = tf.Graph()
with graph.as_default():
    
    tf_r = tf.placeholder(tf.float32,shape=(r_delay, ), name="tf_r")
    tf_u_offline = tf.placeholder(tf.float32,shape=(u_delay, ), name="tf_u_offline")
    tf_u_online = tf.placeholder(tf.float32,shape=(u_delay-1, ), name="tf_u_online")
    tf_y_p = tf.placeholder(tf.float32,shape=(y_p_delay, ), name="tf_y_p")
    tf_e = tf.placeholder(tf.float32,shape=(), name="tf_e")
    
    tf_y_target = tf.placeholder(tf.float32,shape=(), name="tf_y_target")
    tf_r_target = tf.placeholder(tf.float32,shape=(), name="tf_r_target")
    tf_lr = tf.placeholder(tf.float32, shape=(), name="tf_lr")

    def init_weights(layers, weights, biases, output_dim, bias_ones=False):         
        for i in range(len(layers)):
            if i < len(layers) - 1:
                weights.append(tf.Variable(tf.truncated_normal([layers[i], layers[i+1]])))
                if bias_ones:
                    biases.append(tf.Variable(tf.ones([layers[i+1]])))
                else:
                    biases.append(tf.Variable(tf.zeros([layers[i+1]])))

            else:
                weights.append(tf.Variable(tf.truncated_normal([layers[i], output_dim])))
                if bias_ones:
                    biases.append(tf.Variable(tf.ones([output_dim])))
                else:
                    biases.append(tf.Variable(tf.zeros([output_dim])))
    # Model NN    
    #include input
    m_layers = [u_delay + y_p_delay + r_delay, 20, 10, 5, 3]


    m_weights = []
    m_biases = []
    init_weights(m_layers, m_weights, m_biases, 1)

        
    # PID param NN
    #include input
    p_layers = [1 + y_p_delay, 8, 5]

    p_weights = []
    p_biases = []

    
    init_weights(p_layers, p_weights, p_biases, 3, bias_ones=True)


    def forward_prop(input_value, layers, weights, biases):
        for i in range(len(layers)):  
            if i == 0:
                z = tf.matmul(input_value, weights[i]) + biases[i]
            else:
                z = tf.matmul(a, weights[i]) + biases[i]
            if i < len(layers) - 1:
                a = tf.nn.leaky_relu(z)
        return z
        
    #Offline Training
    # Forward prop model offline
    offline_model_input = tf.concat([tf.reshape(tf_r, (1, r_delay)), tf.reshape(tf_u_offline, (1, u_delay)), 
                            tf.reshape(tf_y_p, (1, y_p_delay))], axis=1)
    
    y_model_offline = forward_prop(offline_model_input, m_layers, m_weights, m_biases)
    
    model_var_list = m_weights + m_biases
    model_loss = tf.losses.mean_squared_error(tf_y_target, tf.reshape(y_model_offline, ()))
    model_opt = tf.train.AdamOptimizer(tf_lr).minimize(model_loss, var_list=model_var_list)
    
    #Online Training
    # Forward prop PID
    pid_input = tf.concat([tf.reshape(tf_e, (1, 1)), tf.reshape(tf_y_p, (1, y_p_delay))], axis=1)
    
    pid_gains = forward_prop(pid_input, p_layers, p_weights, p_biases)
    Kp, Ki, Kd = tf.split(pid_gains, 3, axis=1)
    
    Kp_abs = tf.math.abs(Kp)
    Ki_abs = tf.math.abs(Ki)
    Kd_abs = tf.math.abs(Kd)
    #Get signal without changing P I and D terms (may be called multiple times in a time step)
    u_nn = tf_gen_u(Kp_abs, Ki_abs, Kd_abs)
#     u_nn = tf_gen_u(Kp, Ki, Kd)
    
    # Forward prop model online
    online_model_input = tf.concat([tf.reshape(tf_r, (1, r_delay)), tf.reshape(tf_u_online, (1, u_delay-1)), 
                                    u_nn, tf.reshape(tf_y_p, (1, y_p_delay))], axis=1)
                
    y_model_online = forward_prop(online_model_input, m_layers, m_weights, m_biases)

    pid_var_list = p_weights + p_biases
    pid_loss = tf.losses.mean_squared_error(tf_r_target, tf.reshape(y_model_online, ()))
    pid_opt = tf.train.AdamOptimizer(tf_lr).minimize(pid_loss, var_list=pid_var_list)
```

```python
r = r_combo
offline_loss = []
online_loss = []


def get_lr_offline(epoch):
    return 0.01*np.e**(-0.1*epoch)

def get_lr_online(epoch):
    return 0.01*np.e**(-0.1*epoch)

def gen_signal():
    print(pid.Kp, pid.Ki, pid.Kd)

    y_0 = 0
    t_prev = 0
    y_p = np.zeros(steps)
    u = np.zeros(steps)
    for i in range(steps):
        e = r[i] - y_0

        t_cur = t[i]
        pid.set_state(e, t_cur)
        u[i] = pid.update()
        # Get plant output
        y_p[i] = plant(y_0, t_prev, t_cur, u)

        y_0 = y_p[i]
        t_prev = t[i]
    return y_p, u

def reset_online():
    y_p = np.zeros(steps)
    u = np.zeros(steps)
    y_0 = 0
    t_prev = 0
    pid.reset()
    return y_p, u, y_0, t_prev

with tf.Session(graph=graph) as session:
    tf.global_variables_initializer().run()

    def offline_training(r, epochs):
        #Generate signal
        y_p, u = gen_signal()
        y_m = np.zeros(steps)

        #Offline Training
        for j in range(epochs):
            print("Epoch", j)
            lr = get_lr_offline(j)
            print("Lr:", lr)
            epoch_loss = []
            for i in range(steps):
                #Get r_delay number of past r values (including current one)
                if i+1 >= r_delay:
                    r_delayed = r[i+1-r_delay:i+1]
                else:
                    continue
                #Get u_delay number of past u values (including current one)
                if i+1 >= u_delay:
                    u_delayed = u[i+1-u_delay:i+1]
                else:
                    continue
                #Get y_delay number of past y plant values (not including current y value)
                if i >= y_p_delay:
                    y_p_delayed = y_p[i-y_p_delay:i]
                else:
                    continue
                # Train model on plant output

                feed_dict = {tf_r:r_delayed, tf_u_offline:u_delayed, tf_y_p:y_p_delayed,
                             tf_y_target:y_p[i], tf_lr:lr}
                _, l, y_m[i] = session.run([model_opt, model_loss, y_model_offline], feed_dict=feed_dict)
                offline_loss.append(l)
                epoch_loss.append(l)
            print("Offline, avg loss:", sum(epoch_loss)/len(epoch_loss))
        return y_p, y_m, epoch_loss
    
    def online_training(r, epochs, test=False):
        for j in range(epochs):
            print("Epoch", j)
            y_p, u, y_0, t_prev = reset_online()
            y_m = np.zeros(steps)
            Kp_output = np.zeros(steps)
            Ki_output = np.zeros(steps)
            Kd_output = np.zeros(steps)
            lr = get_lr_online(j)
            print("Lr:", lr)
            epoch_loss = []
            for i in range(steps):  
                e = r[i] - y_0
                t_cur = t[i]
                pid.set_state(e, t_cur)

                #Get r_delay number of past r values (including current one)
                if i+1 >= r_delay:
                    r_delayed = r[i+1-r_delay:i+1]
                else:
                    continue
                #Get u_delay-1 number of past u values (not including current time step)
                if i >= u_delay-1:
                    u_delayed = u[i-(u_delay-1):i]
                else:
                    continue
                #Get y_delay number of past y values (not including current y value)
                if i >= y_p_delay:
                    y_delayed = y_p[i-y_p_delay:i]
                else:
                    continue

                # Get get PID parameters and get control signal
                # Send to model to train PID param NN
                if test:
                    feed_dict = {tf_e:e, tf_r:r_delayed, tf_u_online:u_delayed, tf_y_p:y_delayed, 
                                 tf_r_target:r[i]}
                    l, y_m[i] = session.run([pid_loss, y_model_online], feed_dict=feed_dict)
                else:
                    feed_dict = {tf_e:e, tf_r:r_delayed, tf_u_online:u_delayed, tf_y_p:y_delayed, 
                                 tf_r_target:r[i], tf_lr:lr}
                    _, l, y_m[i] = session.run([pid_opt,pid_loss, y_model_online],feed_dict=feed_dict)
                    
                online_loss.append(l)
                epoch_loss.append(l)

                u[i] = pid.update()
                y_p[i] = plant(y_0, t_prev, t_cur, u)
                y_0 = y_p[i]
                t_prev = t[i]
                
                Kp_output[i] = pid.Kp
                Ki_output[i] = pid.Ki
                Kd_output[i] = pid.Kd

            print("Online, avg loss:", sum(epoch_loss)/len(epoch_loss))

        return y_p, y_m, epoch_loss, Kp_output, Ki_output, Kd_output

    y_p, y_m, _ = offline_training(r, 30)
    
        
    plt.plot(t, y_p)
    plt.plot(t, r)
    plt.plot(t, y_m)
    plt.title('Tracking of model to plant after offline training')
    plt.legend(('Plant', 'Command', 'Model'))
    plt.show()
               
    for e in range(10):
        online_training(r, 1)
        offline_training(r, 1)
        
    
    plt.plot(t, y_p)
    plt.plot(t, r)
    plt.plot(t, y_m)
    plt.title('Tracking of model to plant after online training epoch')
    plt.legend(('Plant', 'Command', 'Model'))
    plt.show()
    
    y_p, y_m, test_loss, Kp_output, Ki_output, Kd_output = online_training(r, 1, test=True)
```
