```python
class PID_Neural_Interface:
    def __init__(self):
        self.pid = PID(1, 1, 1)
        self.e = 0
        self.t = 0
        self.y_prev = 0
        self.P = 0
        self.I = 0
        self.D = 0
        self.PID_tensor = tf.constant([[self.P], [self.I], [self.D]], dtype=tf.float32)
        
    def set_state(self, e, t, y_prev):
        self.e = e
        self.t = t
        self.y_prev = y_prev
    
    def catchup_pid_values(self):
        self.P = self.pid.P
        self.I = self.pid.I
        self.D = self.pid.D
    
    def update_stateless(self, x): #Generate command without changing the state of the controller
        delta_time = self.t - self.pid.t_prev
        delta_error =  self.e - self.pid.e_prev
        
        self.P = self.e
    
        self.I = self.pid.I + self.e * delta_time

        if delta_time > 0:
            self.D = delta_error / delta_time
        else:
            self.D = 0
        self.PID_tensor = tf.constant([[self.P], [self.I], [self.D]], dtype=tf.float32)

        return tf.matmul(x, self.PID_tensor) * math.pi/180 + self.y_prev #command is added to previous angle
```

```python
class PID_Layer(Layer):
    def __init__(self, op_function, **kwargs):
        self.op_function = op_function
        super(PID_Layer, self).__init__(**kwargs)

    def build(self, input_shape):
        super(PID_Layer, self).build(input_shape)  # Be sure to call this at the end

    def call(self, x):
        #Logic
        return self.op_function(x)

    def compute_output_shape(self, input_shape):
        return (1, 1)
    
class Abs_Layer(Layer):
    def __init__(self, **kwargs):
        super(Abs_Layer, self).__init__(**kwargs)

    def build(self, input_shape):
        super(Abs_Layer, self).build(input_shape)  # Be sure to call this at the end

    def call(self, x):
        #Logic
        return K.abs(x)

    
```

```python
pid_interface = PID_Neural_Interface()
#https://stackoverflow.com/questions/56657993/how-to-create-a-keras-layer-with-a-custom-gradient-in-tf2-0
@tf.custom_gradient
def pid_op(x):
    u = pid_interface.update_stateless(x)

    def pid_grad(dy):
        grad = tf.multiply(dy, pid_interface.PID_tensor) * math.pi/180
        grad = tf.reshape(grad, (-1, 3))
        return grad
    return u, pid_grad
```

```python
class PIDTuner:
    def __init__(self):
        plant_weights = pkl.load(open('Model Free Approaches/lstm_weights.pkl', 'rb'))
        num_inputs = 1
        num_outputs = 1
        plant_nn = Sequential()
        plant_nn.add(LSTM(16, batch_input_shape=(1, None, num_inputs), return_sequences=True, stateful=True))
        plant_nn.add(Dense(num_outputs))
        plant_nn.layers[0].set_weights(plant_weights[0])
        plant_nn.layers[1].set_weights(plant_weights[1])

        self.pid_nn = Sequential()
        self.pid_nn.add(Dense(4, input_shape=(2,), activation='relu'))
        self.pid_nn.add(Dense(3))
        self.pid_nn.add(Abs_Layer(name="PID_params"))
        self.pid_nn.add(PID_Layer(pid_op, name="u_value"))
        self.pid_nn.add(Reshape((1, 1)))
        self.pid_nn.add(plant_nn)
        self.pid_nn.layers[5].trainable=False
        opt = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, amsgrad=False)
        self.pid_nn.compile(opt, loss='mse')
        print(self.pid_nn.summary())

        
    def predict(self, T, y0, r, d):
        self.pid_nn.reset_states()
        pid_interface.pid.reset()
        y = np.zeros(len(T))
        angle = np.zeros(len(T))
        u = np.zeros(len(T))
        y[0] = y0[1]
        angle[0] = y[0] * 180/math.pi + 180 + d[0]
        
        gains = np.zeros(shape=(len(T), 3))
        
        X = np.zeros(shape=(len(T), 2))
            
        pid_output_model = Model(inputs=self.pid_nn.input, outputs=self.pid_nn.get_layer("PID_params").output)
        u_output_model = Model(inputs=self.pid_nn.input, outputs=self.pid_nn.get_layer("u_value").output)

        for i in range(1, len(T)):

            e = r[i] - angle[i-1]
            e_norm = e/180
            y_prev_norm = (y[i-1]-(-math.pi))/(math.pi-(-math.pi))
            
            X[i, 0] = e_norm
            X[i, 1] = y_prev_norm
            
            pid_nn_input = np.array([X[i, :]])
            
            #Set the variables needed for the stateless PID controller for the neural network to use the PID
            pid_interface.set_state(e, T[i], y[i-1])
            y[i] = self.pid_nn.predict(pid_nn_input, batch_size=1)
#             print(u_output_model.predict(pid_nn_input, batch_size=1))
            
            gains[i, :] = pid_output_model.predict(pid_nn_input, batch_size=1)[0]
            
            #Actually update the state of the PID controller
            pid_interface.pid.update_gains(gains[i, 0], gains[i, 1], gains[i, 2])
            u[i] = pid_interface.pid.update(e, T[i]) * math.pi/180 + y[i-1]
            
            #Catchup the stateless pid controller with the actual one
            pid_interface.catchup_pid_values() 
            
            angle[i] = y[i] * 180/math.pi + 180 + d[i]
            
        return X, y, angle, u, gains
    
#     def train(self, X, r, d):
#         self.pid_nn.reset_states()
#         pid_interface.pid.reset()
#         y_label = (r - d - 180) * math.pi/180

#         for i in range(len(X)):
#             hist = self.pid_nn.fit(np.array(X[i, :]), np.array(y_label[i]), batch_size=1)
#         return hist.history['loss']
```

```python
rot_tuner = PIDTuner()
```
