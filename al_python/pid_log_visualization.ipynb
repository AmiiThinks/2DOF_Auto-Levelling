```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
pid_log = pd.read_csv('pidLog.txt', sep=",", header=None)
```

```python
pid_log.columns = ['time', 'phi_prev', 'setpoint_phi_prev', 'theta_prev', 'setpoint_theta_prev', 
                   'rot_pos_prev', 'rot_vel_prev', 
                   'flex_pos_prev', 'flex_vel_prev',
                   'gx_prev', 'gy_prev', 'gy_prime_prev', 'gz_prev',
                   'phi_pid', 'rot_pid', 'theta_pid', 'flex_pid',
                   'phi_new', 'setpoint_phi_new', 'theta_new', 'setpoint_theta_new', 
                   'rot_pos_new', 'rot_vel_new',
                   'flex_pos_new', 'flex_vel_new',
                   'gx_new', 'gy_new', 'gy_prime_new', 'gz_new']
```

```python
plt.plot(pid_log['phi_prev'])
plt.plot(pid_log['setpoint_phi_prev'])
plt.legend(['phi', 'setpoint'])
plt.show()
```

```python
plt.plot(pid_log['rot_pos_prev'])
plt.plot(pid_log['flex_pos_prev'])
plt.legend(['rot pos', 'flex pos'])
plt.show()
```
