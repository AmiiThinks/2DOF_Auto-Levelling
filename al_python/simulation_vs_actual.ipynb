```python
import numpy as np
from control import *
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
def deg2rad(deg):
    return deg * math.pi / 180
```

```python
dt = 0.005 #Discrete timestep (s)

R = 8.3 #Motor resistance (ohms)
L = 2.03e-3 #Motor inductance (H)
gear_ratio = 193 #Motor gear ratio
gear_eff = 0.836 #Motor gear efficiency
K_w = 93.1 #Speed constant (rad/V)
K_t = 0.0107 #Torque constant (Nm/A)
J_m = 8.68e-8 #Motor moment of inertia (kgm^2)
b_m = 8.87e-8 #Motor friction (Nms)

P_m = 32 #Internal motor P value
I_m = 0 #Internal motor I value
D_m = 0 #Internal motor D value
Vs = 12 #Source voltage V

Kp_m = P_m/8 * 2048 * Vs/(511*math.pi) #Motor Kp
Ki_m = 1000 * I_m / 2048 * 2048 * Vs/(511*math.pi) #Motor Ki
Kd_m = 4 * D_m / 1000 * 2048 * Vs/(511*math.pi) #Motor Kd

m = 0.077 #Servo mass (kg)
b = 0.0355 #Servo width (m)
h = 0.0244 #Servo height (m)
d = 0.033 + 0.012 #Servo depth (m)
theta = 180
bracket_L = 0.03 #Bracket lenght (m)
gripper_com = 0.03 #Centre of mass of gripper wrt where bracket connects

def calcJrot(theta):
    return m * ((b**2 + h**2)/12 + (bracket_L + gripper_com)**2 * math.sin(deg2rad(180-theta))**2)

J_flex = m * (d**2/3 + d * bracket_L + bracket_L**2)

#Simplified tf (eq 31)
c1 = gear_ratio * gear_eff * K_t
c2 = gear_ratio**2 * gear_eff
num = [c1 * Kd_m, c1 * Kp_m, c1 * Ki_m]
d2 = b_m * c2 * R + K_t * c2 / K_w + c1 * Kd_m
d3 = c1 * Kp_m
d4 = c1 * Ki_m

episode = 5
steps = int(episode/dt)

interval = 500
r_steps = steps//interval

r_sq = np.zeros(steps)
up = 0

for i in range(r_steps):
    if i < r_steps-1:
        r_sq[i * interval:(i+1)*interval] = up
    else:
        r_sq[i * interval:] = up
    if up > 0:
        up = -math.pi
    else:
        up = math.pi

r_sin = np.sin(2*np.linspace(0, episode, steps))


r_step = np.zeros(shape=(steps,))
r_step[0:100] = 0
r_step[100:] = math.pi
```

# Effect of flexion position on rotation servo response

```python
u = r_step
T = np.arange(0, 5, dt)

for th in [180, 150, 120, 90]:
    J_rot = calcJrot(th)
    print(J_rot)
    d1 = R * (J_rot + J_m * c2)
    den = [d1, d2, d3, d4]

    sys = tf(num, den)
    Tout, y, xout = forced_response(sys, T, u)
    plt.plot(Tout, y)
plt.plot(T, u)
plt.legend(['180', '150', '120', '90', 'u'])
plt.show()
```

# Comparing simulation to actual

```python
def compare_sim_actual(file, th):
    log = pd.read_csv(file, sep=",")
    grouped = log.groupby(['freq'])

    w_actual= []
    mag_actual = []
    for freq in grouped:
        df = freq[1]
    #     plt.plot(df['sig'])
    #     plt.plot(df['pos'])
    #     plt.legend(['signal', 'position'])
    #     plt.title(str(freq[0]) + " Hz")
    #     plt.show()

        w_actual.append(freq[0]*2*math.pi)

        gain = (df['pos'].max() - df['pos'].min())/(df['sig'].max() - df['sig'].min())
        mag_actual.append(20*np.log(gain))

    if th is not None:
        J_rot = calcJrot(th)
        d1 = R * (J_rot + J_m * c2)
    else:
        d1 = R * (J_flex + J_m * c2)
    den = [d1, d2, d3, d4]
    sys = tf(num, den)
    mag, phase, w = bode_plot(sys, omega=w_actual, Plot=False)
    plt.semilogx(w, 20 * np.log(mag))    # Bode magnitude plot
    plt.semilogx(w_actual, mag_actual)
    plt.xlabel('Frequency (rads/s)')
    plt.ylabel('mag (dB)')
    plt.legend(['sim', 'actual'])

    plt.show()

    T = np.linspace(0, 5, steps)

    for freq in grouped:
        df = freq[1]
        sys = tf(num, den)

        u = df['sig'] / 11.3611111111111111111111 * math.pi / 180 #Ticks to degree to radians
        u = u - math.pi
        Tout, y, xout = forced_response(sys, T, u)

        y_ticks = (y + math.pi) * 180 / math.pi * 11.3611111111111111111111
        plt.plot(T, df['sig'])
        plt.plot(T, y_ticks)
        plt.plot(T, df['pos'])
        plt.legend(['command', 'simulation', 'actual'])
        plt.title(str(freq[0] * 2 * math.pi) + " rad/sec")
        plt.show()
```

```python
#Rotation
compare_sim_actual("rot_log.txt", 180)
```

```python
#Rotation with flexion up
compare_sim_actual("rot_flex_up_log.txt", 90)
```

```python
#Flexion (affected by gravity)
compare_sim_actual("flex_log.txt", None)
```
